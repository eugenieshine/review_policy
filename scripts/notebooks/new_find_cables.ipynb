```python
import json
import pandas as pd
import gzip
```

```python
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l.decode('utf-8'))

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df = getDF('/Users/yujinglai/Dropbox/Eugenie/data/new_amazon_cell_phone_accessories/meta_Cell_Phones_and_Accessories.json.gz')
```

```python
# Drop irrelevant columns
df.drop(['image', 'feature', 'also_buy', 'also_view', 'similar_item', 'date', 'details', 'tech1', 'tech2', 'fit'], axis=1, inplace=True)
```

```python
# Drop records with nan in category
df = df.dropna(subset=['category'])
```

```python
# Find records with Cables in category
df = df.loc[df['category'].apply(lambda x: 'Cables' in x)]
```

```python
# Number of records with 'Cables' in category
df.shape
```

```python
# Get records whose rank is list
# These are headphone adapters
df_lsRank = df.loc[df['rank'].apply(lambda x: type(x) != str)]
```

```python
# Drop records with rank as list
df = df.loc[df['rank'].apply(lambda x: type(x) == str)]
```

```python
# Get rank
df['rank'] = df['rank'].apply(lambda x: x.split('i')[0])
```

```python
# Trim special characters
df['rank'] = df['rank'].apply(lambda x: x.replace(">#",""))
```

```python
# Parse str to int
df['rank'] = df['rank'].apply(lambda x: int(float(int(x.replace(',', '')))))
```

```python
df.shape
```

```python
def findCables(x):
    return 'Charger' in x or 'Charging' in x or 'charging' in x or 'charger' in x or 'USB' in x or 'usb' in x;
```

```python
# Find cables based on key words
df = df.loc[df['title'].apply(findCables)]
```

```python
# Get the top 200 records
df_top200 = df.sort_values('rank').head(200)
```
