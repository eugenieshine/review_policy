### To test textblob package for sentiment analysis in an affortable way

```python
from textblob import TextBlob
```

```python
import pandas as pd
import numpy as np
```

```python
def detect_polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
DIR_PATH = '../../../../../Dropbox/Eugenie/data/processed_reviewMeta/'
DIR_PATH_Mc = '../../../../../Dropbox/Eugenie/data/processed_julian_amazon_data/sentiment/'
```

```python
df = pd.read_csv(DIR_PATH+'merged-reviews.csv')
```

```python
df.columns
```

```python
df.shape
```

```python
sample = df[['rating','positive','text']].sample(1000)
```

```python
sample.rename(columns={'positive':'AWS_polarity'},inplace=True)
```

```python
sample['textblob_polarity'] = sample.text.apply(detect_polarity)
```

```python
#Normalized Data
sample['normalized_aws'] = sample['positive'].apply(lambda x: 2*(x-min(sample['positive']))/(max(sample['positive'])-min(sample['positive']))-1)
sample['normalized_polarity'] = sample['textblob_polarity'].apply(lambda x:(x-min(sample['textblob_polarity']))/(max(sample['textblob_polarity'])-min(sample['textblob_polarity'])))
```

```python
sample.head()
```

### nltk package

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
def nltk_polarity(text):
    return sid.polarity_scores(text)['compound']
```

```python
sample['nltk_polarity'] = sample.text.apply(nltk_polarity)
```

### flair package – NNs

```python
import sklearn; print("Scikit-Learn", sklearn.__version__)
```

```python
import flair
flair_sentiment = flair.models.TextClassifier.load('en-sentiment')
```

```python
def flair_polarity(text):
    s = flair.data.Sentence(text)
    flair_sentiment.predict(s)
    total_sentiment = s.labels[0].to_dict()
    if total_sentiment['value'] == 'NEGATIVE':
        return -total_sentiment['confidence']
    return total_sentiment['confidence']
```

```python
sample['flair_polarity'] = sample['text'].apply(flair_polarity)
```

```python
def take_median(x):
    print(x.to_dict())
```

```python
sample['median_polarity'] = sample[['textblob_polarity','nltk_polarity','flair_polarity']].median(axis=1)
```

```python
sample.to_csv('sample')
```

Comparison

```python
import itertools
```

```python
p_cols = ['textblob_polarity','nltk_polarity','flair_polarity','median_polarity']
```

```python
plt.figure(figsize=(18,11))
img_num = 0
for img_num, i in enumerate(itertools.combinations(p_cols, 2)):
    plt.subplot(2, 3, img_num+1)
    x = sample[i[0]].values
    y = sample[i[1]].values
    plt.scatter(x, y, s=3, c='b')
    plt.plot([-1, 1], [-1, 1], c='r', linewidth=0.75, zorder=0)
    plt.grid()
    plt.xlabel(i[0])
    plt.ylabel(i[1])
    plt.title('Sentiment Score')
    #img_num = img_num + 1
plt.savefig('plots/compare_sentiment_packages.png')
plt.show()
```

```python
plt.figure(figsize=(18,10))
for img_num, p in enumerate(p_cols):
    pos = []
    neg = []
    plt.subplot(2, 3, img_num+1)
    for group in sample.groupby('rating', as_index=True):
        df_group = group[1]
        pos.append(df_group.loc[df_group[p]>=0].shape[0])
        neg.append(df_group.loc[df_group[p]<0].shape[0])
    index = np.arange(1,6)
    bar_width = 0.35
    plt.bar(index, pos, bar_width, alpha=0.5, label='Positive')
    plt.bar(index+bar_width, neg, bar_width, alpha=0.5, color='r', label='Negative')
    plt.grid()
    plt.legend()
    plt.xlabel('rating')
    plt.ylabel(p)
plt.savefig('plots/sentiment_vs_rating.png')
plt.show()
```
