```python
reviews_df = pd.DataFrame(columns=['line', 'sentiment_class', 'mixed', 'negative', 'neutral', 'positive'])
```

```python
for r in reviews:
    li = []
    ## ensure it's a valid record
    if len(r) > 1:
        ## turn to dict structure
        r_dict = json.loads(r)
        
        ## check error message
        if 'Sentiment' in r_dict:
            li.append(r_dict['Line'])
            li.append(r_dict['Sentiment'])
            print(r_dict['Line'])
            ## turn sentiment scores to a list
            r_scores = list(pd.DataFrame.from_dict(r_dict['SentimentScore'], orient='index')[0])
            li.extend(r_scores)
            
            ## add a new row with record data
            reviews_df[r_dict['Line']] = li
```

```python
## this is a list of dictionaries
review_dict_li = []

for r in reviews:
    ## ensure it's a valid record
    if len(r) > 1:
        ## turn to dict structure
        r_dict = json.loads(r)
        review_dict_li.append(r_dict)
```
